```python
im = Image.open(r"C:\Users\OrekhovNA\Pictures\pygame\pacman\background_old.png") 

pixels = list(im.getdata())
width, height = im.size
pixels = [pixels[i * width:(i + 1) * width] for i in range(height)]
```

```python
borders = list()
borders_by_rows = dict()
borders_by_cols = dict()
for y, row in enumerate(pixels):
    for x, val in enumerate(row):
        if val[0] != 255 or val[1] != 255 or val[2] != 255 or val[3] != 255:
            borders.append((x, y))
            
            if y not in borders_by_rows:
                borders_by_rows[y] = list()
            if x not in borders_by_cols:
                borders_by_cols[x] = list()
            
            borders_by_rows[y].append(x)
            borders_by_cols[x].append(y)
            
```
